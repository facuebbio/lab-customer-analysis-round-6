```python
## Checking and Cleaning OUTLIERS of "monthly premium auto" and "total claim amount".
```

```python
## MONTHLY PREMIUM AUTO.

# To deal with the outliers we have to:

# Check if there are outliers
sns.boxplot(x= dataset["monthly premium auto"])
plt.show()

# We have a huge amount of Outliers in this feature

# we are gonna clean the outliers

## IQR = 03-Q1
IQR = np.percentile(dataset["monthly premium auto"], 75) - np.percentile(dataset["monthly premium auto"], 25)
## upper_limit = Q3 + 1.5 *IQR
## lower_limit = Q1 - 1.5 * IQR
upper_limit = np.percentile(dataset["monthly premium auto"], 75) + 1.5 *IQR
lower_limit = np.percentile(dataset["monthly premium auto"], 25) - 1.5 *IQR

mask_cleaned = (dataset["monthly premium auto"] < upper_limit) & (dataset["monthly premium auto"] > lower_limit)
mask_outliers = (dataset["monthly premium auto"] > upper_limit) | (dataset["monthly premium auto"] < lower_limit)

dataset_no_outliers_monthly = dataset[mask_cleaned]
len(dataset_no_outliers_monthly)/len(dataset)
```

```python
## PLOTTING CLEANED (NO OUTLIERS) FEATURE
```

```python
sns.boxplot(dataset_no_outliers_monthly["monthly premium auto"])
```

```python
sns.distplot(dataset_no_outliers_monthly["monthly premium auto"])
```

```python
## TOTAL CLAIM AMOUNT

# To deal with the outliers we have to:

# Check if there are outliers
sns.boxplot(x= dataset["total claim amount"])
plt.show()

# We have a huge amount of Outliers in this feature as well

# we are gonna clean the outliers

## IQR = 03-Q1
IQR = np.percentile(dataset["total claim amount"], 75) - np.percentile(dataset["total claim amount"], 25)
## upper_limit = Q3 + 1.5 *IQR
## lower_limit = Q1 - 1.5 * IQR
upper_limit = np.percentile(dataset["total claim amount"], 75) + 1.5 *IQR
lower_limit = np.percentile(dataset["total claim amount"], 25) - 1.5 *IQR

mask_cleaned = (dataset["total claim amount"] < upper_limit) & (dataset["total claim amount"] > lower_limit)
mask_outliers = (dataset["total claim amount"] > upper_limit) | (dataset["total claim amount"] < lower_limit)

dataset_no_outliers_claim = dataset[mask_cleaned]
len(dataset_no_outliers_claim)/len(dataset)
```

```python
## Processing Data
## X-y SPLIT

dataset["total claim amount"].unique()
```

```python
display(dataset)

X = dataset.iloc[:, : -1] # explanatory variables
y = dataset.iloc[:,-1] #  target feature
```

```python
# dummify the categorical variables

X = pd.get_dummies(X, drop_first=True)
X
```

```python
X
```

```python
y
```

```python
# SPLITTING

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
display(X_train)
X_test
```

```python
len(X_train)/len(X)
```

```python
# SCALING THE NUMERICAL VARIABLES
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
# fit transform the data with the scaler

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
## I will create the Linear Regression with sklearn

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train, y_train)
```

```python
regressor.coef_ # get the coefficients of the model
```

```python
regressor.intercept_ #b0
```

```python
# performances
print("train set score:", regressor.score(X_train, y_train))
print("test set score:", regressor.score(X_test, y_test))
# it is a bad model, underfitting model.
```

```python
predictions_trainset = regressor.predict(X_train)
predictions_testset = regressor.predict(X_test)
```

```python
# MAE, MSE, R2

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rmse = mean_squared_error(y_train,predictions_trainset)
mae = mean_absolute_error(y_train,predictions_trainset)
r2 = r2_score(y_train,predictions_trainset)
print(rmse, mae, r2)
```

```python
rmse = mean_squared_error(y_test,predictions_testset)
mae = mean_absolute_error(y_test,predictions_testset)
r2 = r2_score(y_test,predictions_testset)
print(rmse, mae, r2)
```
